```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt
from plot_tools import plot_response
```

$P(s) = \frac{2}{s + 3}$

$C(s) = K_p$

```python
kp = 3


P = ct.tf([2], [1, 3])
C = ct.tf([kp], [1])

display(P)
```

```python
Gyr = ct.feedback(P*C, 1)
display(Gyr)
```

```python
dt = 0.01
T = 10

t = np.arange(0, T, dt)

r = np.sign(np.sin(t))
outy = ct.forced_response(Gyr, t, r)
```

```python
fig, ax = plot_response(t, r, outy.y[0]);
plt.show()
```

```python

Gur = ct.feedback(C, P)
outu = ct.forced_response(Gur, t, r)
fig, ax = plot_response(t, r, outu.y[0]);
plt.show()
```
