```python
instr_mem = rv_core.instr_mem
data_mem = rv_core.data_mem
```

```python
with open ('data128.dat', "r") as data:
    data_lines = data.readlines()
    
with open ('code128.dat', "r") as instr:
    instr_lines = instr.readlines()
    
instr_pt = 0
data_pt = 0
    
from pynq import MMIO
reboot = MMIO(0x43C0_0000,0xFFFF)

reboot.write(0x10,1)

for data_line in data_lines:
    data_mem.write(data_pt,int(data_line,16))
    data_pt = data_pt + 4
    
for instr_line in instr_lines:
    instr_mem.write(instr_pt,int(instr_line,16))
    instr_pt = instr_pt + 4

    
instr_pt = 0
data_pt = 0

reboot.write(0x10,0)
```

```python
data_pt = 0
line = 0
while data_pt < 2049:
    print(line, hex(data_mem.read(data_pt)))
    data_pt += 4
    line += 1
```
