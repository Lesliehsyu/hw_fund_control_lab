```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt
from plot_tools import plot_response
```

$P(s) = \frac{25}{s^2 + 6s + 25}$

$C(s) = K_p + K_d s$

```python
kp = 3
kd = 0


P = ct.tf([25], [1, 6, 25])
C = ct.tf([kd, kp], [1])

display(P)
display(C)
```

```python
Gyr = ct.feedback(P*C)
display(Gyr)
```

```python
dt = 0.01
T = 10

t = np.arange(0, T, dt)

r = np.sign(np.sin(t))
outy = ct.forced_response(P, t, r)
```

```python
fig, ax = plot_response(t, r, outy.y[0]);
plt.show()
```

```python

Gur = ct.feedback(C, P)
display(Gur)
outu = ct.forced_response(Gur, t, r)
fig, ax = plot_response(t, r, outu.y[0]);
plt.show()
```
